### 3. summary

- tokenizer 轻易不会将一个词处理为 `[UNK] (100)`
- 基于词汇表，tokenize, encode, decode 一体
    - tokenize：word => token(s)，将word尽可能地映射为 vocab 中的 keys
    - encode: token => id
    - decode: id => token => word
        - encode 完了之后也不是终点（word），decode 还要能很好地将 id 还原，尽可能与输入的 word 对齐；
